Statistics is a branch of mathematics that deals with the collection, analysis, interpretation, presentation, and organization of data.
There are two main branches of statistics: descriptive statistics and inferential statistics.
i.	Descriptive statistics: Is a branch of statistics that focuses on summarizing and presenting data in a meaningful way.

-Mean : It represents the sum of all values in a dataset divided by the number of
values.

-	Median: It is the middle value in a dataset when the values are arranged in ascending order. If there's an even number of values, it's the average of the two middle values

-	Mode: The mode is the value that appears most frequently in a dataset

ii.	Inferential statistics
Inferential statistics is a branch of statistics that involves using data from a sample to make conclusions.


Hypothesis Testing: It involves making inferences about a population based on a
sample

Regression Analysis: This is used to analyze the relationship between two or more
variables



3.Descriptive statistics aim to summarize and describe the main characteristics of a dataset. These techniques are used when you want to understand the basic features of your data without making inferences or generalizations about a larger population.

```python
#replacing commas with empty spaces 
economics.replace(',', '', regex=True, inplace= True)
```

```python
#converting Series of object type to float type
economics['Human Capital Index (HCI) (scale 0-1)'] = economics['Human Capital Index (HCI) (scale 0-1)'].astype("float")
economics['GDP (current US$)current US$constant US$current LCUconstant LCU'] = economics['GDP (current US$)current US$constant US$current LCUconstant LCU'].astype("float")
economics['GDP per capita (current US$)current US$constant US$current LCUconstant LCU'] = economics['GDP per capita (current US$)current US$constant US$current LCUconstant LCU'].astype("float")
economics['GDP growth (annual %)'] = economics['GDP growth (annual %)'].astype("float")
economics['Foreign direct investment, net inflows (% of GDP)'] = economics['Foreign direct investment, net inflows (% of GDP)'].astype("float")
```

```python
economics.info()
```

```python
#imputing missing values with column mean values 
mean_human = economics['Human Capital Index (HCI) (scale 0-1)'].mean()
mean_GDB_current = economics['GDP (current US$)current US$constant US$current LCUconstant LCU'].mean()
mean_GDB_perCapita = economics['GDP per capita (current US$)current US$constant US$current LCUconstant LCU'].mean()
mean_GDB_growth = economics['GDP growth (annual %)'].mean()
mean_foreign = economics['Foreign direct investment, net inflows (% of GDP)'].mean()
economics.fillna({'Human Capital Index (HCI) (scale 0-1)': mean_human,
                              'GDP (current US$)current US$constant US$current LCUconstant LCU': mean_GDB_current,
                             'GDP per capita (current US$)current US$constant US$current LCUconstant LCU':mean_GDB_perCapita,
                             'GDP growth (annual %)':mean_GDB_growth,
                             'Foreign direct investment, net inflows (% of GDP)':mean_foreign}, inplace= True)
```
