# Bivariate Analysis

```python
#comparing Price and Rooms 
sns.scatterplot(y=real_estate.Price, x=real_estate.Rooms, hue=real_estate.Type)
plt.title('RELATIONSHIP BETWEEN PRICE AND ROOM SIZE')
plt.xlabel('Rooms')
plt.ylabel('Price in USD')
```

```python
#relationship between distance and price 
sns.regplot(y=real_estate.Price, x=real_estate.Distance)
plt.title('RELATIONSHIP BETWEEN PRICE AND DISTANCE SIZE')
plt.xlabel('Distance')
plt.ylabel('Price in USD')
```
